```python
import pandas as pd
```

```python
student_info = pd.read_csv('raw_data/studentInfo.csv')
```

```python
student_info.head()
```

```python
student_info['imd_band'] = student_info['imd_band'].replace('10-20', '10-20%')
```

```python
student_info['imd_band'] = student_info['imd_band'].replace('?', 'Unknown')
```

```python
student_info['disability'] = student_info['disability'].map({'Y':True, 'N':False})
```

```python
student_info.info()
```

```python
student_assess = pd.read_csv('raw_data/studentAssessment.csv')
```

```python
student_assess.head()
```

```python
student_assess['score'] = pd.to_numeric(student_assess['score'], errors='coerce')
```

```python
student_assess['is_banked'] = student_assess['is_banked'].astype(bool)
```

```python
student_assess.info()
```

```python
student_vle = pd.read_csv('raw_data/studentVle.csv')
```

```python
student_vle.head()
```

```python
student_vle.info()
```

```python
# student_vle = student_vle.groupby(['code_module', 'code_presentation', 'id_student', 'id_site'])['sum_click'].sum().reset_index(name='total_clicks')
student_vle = student_vle.groupby(['code_module', 'code_presentation', 'id_student'])['id_site'].count().reset_index(name='site_count')
```

```python
assessments = pd.read_csv('raw_data/assessments.csv')
```

```python
assessments.head()
```

```python
assessments.info()
```

```python
assessments['date'] = pd.to_numeric(assessments['date'], errors='coerce')
```

```python
assessments['date'] = assessments['date'].astype('Int64')
```

```python
student_info_clean = student_info
student_info_clean.to_csv('cleaned_data/clean_studentInfo.csv', index=False)
```

```python
student_assess_clean = student_assess
student_assess_clean.to_csv('cleaned_data/clean_studentAssessment.csv', index=False)
```

```python
student_vle_clean = student_vle
student_vle_clean.to_csv('cleaned_data/clean_studentVle.csv', index=False)
```

```python
assessments_clean = assessments
assessments_clean.to_csv('cleaned_data/clean_assessments.csv', index=False)
```
