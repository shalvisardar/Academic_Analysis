```python
import pandas as pd
```

```python
student_assessment_db = pd.read_csv('cleaned_data/clean_studentAssessment.csv')
assessments_db = pd.read_csv('cleaned_data/clean_assessments.csv')
```

```python
assessment_scores_db = pd.merge(student_assessment_db,assessments_db, how='left', on='id_assessment')
assessment_scores_db = assessment_scores_db[
    [
        'code_module',
        'code_presentation',
        'id_student',
        'id_assessment',
        'assessment_type',
        'score'
    ]
]
```

```python
assessment_scores_db.sort_values(by=['id_student', 'id_assessment'], inplace=True)
```

```python
assessment_scores_db.head()
```

```python
assessment_scores_db.info()
```

```python
assessment_pivot_db = assessment_scores_db.pivot_table(index='id_student', columns='assessment_type', values='score', aggfunc='mean')
```

```python
assessment_pivot_db = assessment_pivot_db[['CMA', 'TMA', 'Exam']]
```

```python
assessment_pivot_db.reset_index(inplace=True)
```

```python
assessment_pivot_db.columns.name = None
```

```python
assessment_pivot_db.head()
```

```python
assessment_pivot_db[['CMA', 'TMA', 'Exam']].describe()
```

```python
student_info_db = pd.read_csv('cleaned_data/clean_studentInfo.csv')
student_vle_db = pd.read_csv('cleaned_data/clean_studentVle.csv')
```

```python
student_details_db = pd.merge(student_info_db, student_vle_db, how='left', on=['id_student', 'code_presentation', 'code_module'])
student_details_db = pd.merge(student_details_db, assessment_pivot_db, how='left', on='id_student')
```

```python
student_details_db = student_details_db[
    [
        'id_student',
        'code_module',
        'code_presentation',
        'gender',
        'highest_education',
        'region',
        'imd_band',
        'age_band',
        'disability',
        'num_of_prev_attempts',
        'studied_credits',
        'site_count',
        'CMA',
        'TMA',
        'Exam',
        'final_result'
    ]
]
```
