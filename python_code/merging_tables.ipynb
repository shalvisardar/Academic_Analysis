```python
import pandas as pd
```

```python
student_assessment = pd.read_csv('cleaned_data/clean_studentAssessment.csv')
assessments = pd.read_csv('cleaned_data/clean_assessments.csv')
```

```python
assessment_scores = pd.merge(student_assessment,assessments, how='left', on='id_assessment')
assessment_scores = assessment_scores[
    [
        'code_module',
        'code_presentation',
        'id_student',
        'id_assessment',
        'assessment_type',
        'weight',
        'score'
    ]
]
```

```python
assessment_scores.sort_values(by=['id_student', 'id_assessment'], inplace=True)
```

```python
assessment_scores.head()
```

```python
assessment_scores.info()
```

```python
print(assessment_scores['score'].corr(assessment_scores['weight']))
```

```python
assessment_pivot = assessment_scores.pivot_table(index='id_student', columns='assessment_type', values='score', aggfunc='mean')
```

```python
assessment_pivot = assessment_pivot[['CMA', 'TMA', 'Exam']]
```

```python
assessment_pivot.reset_index(inplace=True)
```

```python
assessment_pivot.columns.name = None
```

```python
assessment_pivot.info()
```

```python
assessment_pivot.head()
```

```python
assessment_pivot[['CMA', 'TMA', 'Exam']].describe()
```

```python
student_info = pd.read_csv('cleaned_data/clean_studentInfo.csv')
student_vle = pd.read_csv('cleaned_data/clean_studentVle.csv')
```

```python
assessment_pivot['id_student'].nunique()
```

```python
student_details = pd.merge(student_info, student_vle, how='left', on=['id_student', 'code_presentation', 'code_module'])
student_all_features = pd.merge(student_details, assessment_pivot, how='left', on='id_student')
```

```python
student_all_features = student_all_features[
    [
        'id_student',
        'code_module',
        'code_presentation',
        'gender',
        'highest_education',
        'imd_band',
        'age_band',
        'disability',
        'num_of_prev_attempts',
        'studied_credits',
        'site_count',
        'CMA',
        'TMA',
        'Exam',
        'final_result'
    ]
]
```
